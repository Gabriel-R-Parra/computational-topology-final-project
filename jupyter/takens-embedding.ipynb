```python
import pandas as pd
import seaborn as sns
from metadata import get_json
sns.set_style("darkgrid")
muscles = get_json("muscles")
```

```python
def takens_indices(tau, d, n_samples):
    def _takens_indices(t):
        return [t + n * tau for n in range(d + 1)]
    return list(filter(lambda x: x[-1] < n_samples, map(_takens_indices, range(n_samples))))
```

```python
data = pd.read_csv("data/02-01-01.csv")[muscles]
```

```python
sns.lineplot(data, legend=False, dashes=False);
```

```python
takens_embeddings = data.values[takens_indices(tau=3, d=2, n_samples=data.shape[0])]
```

```python
takens_vector_embeddings = takens_embeddings[:].reshape(takens_embeddings.shape[0], -1)
```

The variable `takens_vector_embeddings` stores the matrix valued embeddings as a vector using the transformation $x \in \mathbb{R}^{d \times 8} \mapsto x' \in \mathbb{R}^{8d}$
