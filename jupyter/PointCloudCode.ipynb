```python
def embedding(x):
    m = 3
    tau = 5
    # print(x)
    N = len(x)
    M = N - (m - 1) * tau

    return np.column_stack([x[i : i + M] for i in range(0, m * tau, tau)])
```

```python
embedBB = embedding(data["BB"])
print(embedBB.shape)
```

```python
files = os.listdir("data")

print(len(files))
n = 0

# v = files[5]
# print(v[:8])


for file in files:
    # print(file[7])
    if int(file[7]) < 5:
        print(file)
        info = pd.read_csv(f"data/{file}")
        info = info[info.columns[1:9]]

        shortFile = file[:8]
        embedBBBB = embedding(info["BR"])
        # print(embedBBBB)
        np.save(f"pointClouds/{shortFile}BR.npy", embedBBBB)


print("end")

# print(n)
```
